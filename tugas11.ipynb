```python
# Buatlah sebuah DataFrame dari data jumlah produksi sampah berdasarkan Kabupaten/Kota di Jawa Barat.
#  Pastikan kolom-kolomnya menyertakan nama Kabupaten/Kota, jumlah produksi sampah (dalam ton), dan tahun pencatatan."
import pandas as pd

#Baca data dari file excel 
data_jabar = pd.read_excel('disperkim-od_16985_jumlah_produksi_sampah_berdasarkan_kabupatenkota_v3_data.xlsx', sheet_name='data')
data_jabar
```

```python
sampah_jabar= pd.DataFrame(data_jabar, columns=['nama_kabupaten_kota','jumlah_produksi_sampah','tahun'])
sampah_jabar

sampah_jabar.to_csv('sampah_jabar.csv')
sampah_jabar.to_excel('sampah_jabar.xlsx')
```

```python
# NO 2

d16 = {}  # jumlah sampah pada tahun 2016

for i, row in data_jabar.iterrows():
    tahun = row['tahun']
    sampah = row['jumlah_produksi_sampah']

    if tahun == 2016 and tahun not in d16:
        d16[tahun] = sampah
    elif tahun == 2016:
        d16[tahun] += sampah

print('=============== Data Tahun 2016 ===============')
for i, j in d16.items():
    print(f"Jumlah sampah pada tahun {i} adalah {j:.2f} ton")

df_t2016 = pd.DataFrame([
    {'Tahun': i, 'Total/tahun': j}
    for i, j in d16.items()
])

df_t2016.to_csv('df_t2016.csv', index=False)
df_t2016.to_excel('df_t2016.xlsx', index=False)
```

```python
# NO 3
# Jumlahkan Data Pertahun

# Menginisialisasi dictionary untuk menyimpan hasil
thn = {}

# Misalkan df_jabar adalah DataFrame yang telah didefinisikan sebelumnya
for i,j in data_jabar.iterrows():
    tahun = j['tahun']
    total_sampah = j['jumlah_produksi_sampah']

    if tahun not in thn:
        thn[tahun] = total_sampah
    else:
        thn[tahun] += total_sampah

print ('================== Data per Tahun ==================')
for i, j in thn.items():
    print(f'Pada Tahun {i} terdapat : {j:.2f} ton sampah')

df_sampahthn = pd.DataFrame([
    {'Tahun':i, 'Total Sampah' :j}
    for i,j in thn.items()
])

df_sampahthn

df_sampahthn.to_csv('df_sampahthn.csv')
df_sampahthn.to_excel('df_sampahthn.xlsx')
```

```python
# NO 4
# Jumlahkan data per Kota/Kabupaten per tahun
total_kab_kot = {}

for i,j in df_jabar.iterrows():
     daerah = j['nama_kabupaten_kota']
     sampah = j['jumlah_produksi_sampah']
     
     if daerah not in total_kab_kot : 
          total_kab_kot[daerah] = sampah
     else : 
          total_kab_kot[daerah] += sampah
print ('===================== Data Per Daerah =====================')
for i, j in total_kab_kot.items():
     print (f'Pada Daerah{i} terdapat {j:.2f} ton sampah')

df_smphderah = pd.DataFrame([
     {'Nama Kabupaten/Kota' :i, 'Total Sampah' : j}
     for i, j in total_kab_kot.items()
])
df_smphderah
df_sampahthn.to_csv('df_totalkabkot.csv')
df_sampahthn.to_excel('df_totalkabkot.xlsx')
```
